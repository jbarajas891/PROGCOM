```python


import random
import time


def player():

    '''Here the player can read the rules of the game and select one of the "moves" for playing '''

    
    while True:
        player_choice = input('''
**************************************
Selecccione!
Roca
Papel
tijeras
lagarto
Spock
.
**************************************\n
''').lower()

        if player_choice == "reglas":
            print('''
Las tijeras cortan el papel.
El papel cubre la piedra.
La piedra aplasta el lagarto.
El lagarto envenena a Spock.
Spock aplasta las tijeras.
Las tijeras decapitan el lagarto.
El lagarto se come el papel.
El papel refuta a Spock.
Spock vaporiza la piedra.
La piedra aplasta a las tijeras.!!
''')
            time.sleep(3)

        elif player_choice in ["roca", "papel", "tijeras", "lagarto", "spock"]:
            print(f'\n\nSeleccionaste: {player_choice}\n')
            return player_choice 
        else:
            print("\nTa mal vuelva a intentarlo")
            time.sleep(1)

        
def opponent():

    ''' la matrix escogera una opcion aleatorea '''

    print("\nESPERANDO JUAGADA DEL OPONENTE\n")
    time.sleep(1)
    choices = ["roca", "papeL", "tijeras", "lagarto", "spock"]
    opponent_choice = random.choice(choices)
    print(f'\nTu oponente jugó: {opponent_choice.capitalize()}.\n\n')
    return opponent_choice

def resolution(player_choice, opponent_choice):

    

    time.sleep(1)
    winmoves = {
        "roca": ["tijeras", "lagarto"],
        "papel": ["roca", "spock"],
        "tijeras": ["papel", "lagarto"],
        "lagarto": ["papel", "spock"],
        "spock": ["tijeras", "roca"]
        }

    if player_choice == opponent_choice:
        print("empatamos!\n")
        return

 
    elif opponent_choice in winmoves[player_choice]:
        print("ganaste!!\n")
    else:
        print("perdiste!\n")


if __name__ == "__main__":
    print('bienvenido a roca papel o tijera ademas lagarto y spock')

    # para seguir jugando
    retry_choice = None
    while retry_choice != "q":
        player_move = player()
        opponent_move = opponent()
        resolution(player_move, opponent_move)
        retry_choice = input('\nescribe "q" para detener "qq para seguir" ! ').lower()
```
