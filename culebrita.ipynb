```python
from re import X
import turtle
import time
import random

posponer = 0.1

puntaje = 0
maximo = 0
 # config de  la ventana
wn = turtle.Screen()
wn.title('juego de la culebrita')
wn.bgcolor('red')
wn.setup(width= 600,height=600)
wn.tracer(0)
#cabeza de la culebra
cabeza = turtle.Turtle()
cabeza.speed(0)
cabeza.shape('square')
cabeza.color('blue')
cabeza.penup()
cabeza.goto(0,0)
cabeza.direction = 'stop'
#comida
comida = turtle.Turtle()
comida.speed(0)
comida.shape('circle')
comida.color('white')
comida.penup()
comida.goto(0,100)
comida.direction = 'stop'
#texto
texto = turtle.Turtle()
texto.speed(0)
texto.color('green')
texto.penup()
texto.hideturtle()
texto.goto(0,260)
texto.write("puntaje:0           maximo:0", align='center', font=('Courtier',30, 'normal'))



#cuando el cuerpo de la culebra crece
segmentos =[]
#funciones de la culebra
def arriba():
    cabeza.direction = 'Up'
def abajo():
    cabeza.direction = 'Down'
def izquierda():
    cabeza.direction = 'Left'
def derecha():
    cabeza.direction = 'Right'



def mov():
    if cabeza.direction =='Up':
        y = cabeza.ycor()
        cabeza.sety(y + 20)
    if cabeza.direction =='Down':
        y = cabeza.ycor()
        cabeza.sety(y - 20)
    if cabeza.direction =='Left':
        x = cabeza.xcor()
        cabeza.setx(x - 20)
    if cabeza.direction =='Right':
        x = cabeza.xcor()
        cabeza.setx(x + 20)

#para las teclas 
wn.listen()
wn.onkeypress(arriba,'Up')      
wn.onkeypress(abajo,'Down')
wn.onkeypress(izquierda,'Left') 
wn.onkeypress(derecha,'Right')  
while True:
    wn.update()
    #colisiones con los margenes pa perder
    if cabeza.xcor()> 280 or cabeza.xcor() < -280 or cabeza.ycor() > 280 or cabeza.ycor() < -280:
        time.sleep(1)
        cabeza.goto(0,0)
        cabeza.direction = 'stop'

        # esconder segmentos.
        for segmento in segmentos:
            segmento.goto(1000,1000)
        #limpiar segmentos de lista para que quede la cabeza solita
        segmentos.clear()
        #reset puntaje
        puntaje=0
        texto.clear()
        texto.write("puntaje: {}           maximo: {} ".format(puntaje,maximo), align='center', font=('Courtier',30, 'normal'))






    #coliones con la comida
    if cabeza.distance(comida)<20:
        x=random.randint(-280,280)
        y=random.randint(-280,280)
        comida.goto(x,y)
        nuevo_segmento = turtle.Turtle()
        nuevo_segmento.speed(0)
        nuevo_segmento.shape('square')
        nuevo_segmento.color('grey')
        nuevo_segmento.penup()
        nuevo_segmento.goto(0,0)
        segmentos.append( nuevo_segmento)
        #aumentar puntaje
        puntaje += 10
        if puntaje > maximo:
            maximo= puntaje
        texto.clear()
        texto.write("puntaje: {}           maximo: {} ".format(puntaje,maximo), align='center', font=('Courtier',30, 'normal'))

    #mover el cuerpesito de la culebra
    totalSeg = len(segmentos)
    for index in range(totalSeg -1,0,-1):
        x = segmentos[index-1].xcor()
        y = segmentos[index-1].ycor()
        segmentos[index].goto(x,y)
    if totalSeg > 0:
        x = cabeza.xcor()
        y = cabeza.ycor()
        segmentos[0].goto(x,y)



        
    mov()
    #colisones con culebrita
    for segmento in segmentos:
        if segmento.distance(cabeza)<20:
            time.sleep(1)
            cabeza.goto(0,0)
            cabeza.direction = 'stop'
            #esconder segementos
            for segmento in segmentos:
                segmento.goto(1000,1000)
            #limpiar elementos de la lista
            segmentos.clear()
            #reset puntaje
            puntaje=0
            texto.clear()
            texto.write("puntaje: {}           maximo: {} ".format(puntaje,maximo), align='center', font=('Courtier',30, 'normal'))


    time.sleep(posponer)
```
