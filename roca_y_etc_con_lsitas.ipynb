```python


import random
import time


def player():

    

    
    while True:
        player_choice = input('''
**************************************
Selecccione!
Roca
Papel
tijeras
lagarto
Spock
papa
heavy
.
**************************************\n
''').lower()

        if player_choice == "reglas":
            print('''
Las tijeras mata el papel.
El papel mata la roca.
La roca mata el lagarto.
El lagarto mata a Spock.
Spock mata las tijeras.
Las tijeras mata el lagarto.
El lagarto mata el papel.
El papel mata a Spock.
Spock mata la roca.
La roca mata a las tijeras.
heavy mata papa.
heavy mata lagarto.
papa mata a papel.
papa mata a roca
roca mata heavy .
tijera mata heavy.
spock mata a papa.
\ 4_4  /
 \____/


''')
            time.sleep(3)

        elif player_choice in ["roca", "papel", "tijeras","heavy","papa", "lagarto", "spock"]:
            print(f'\n\nSeleccionaste: {player_choice}\n')
            return player_choice 
        else:
            print("\nTa mal vuelva a intentarlo")
            time.sleep(1)

        
def opponent():

    ''' la matrix escogera una opcion aleatorea '''

    print("\nESPERANDO JUAGADA DEL OPONENTE\n")
    time.sleep(1)
    choices = ["roca", "papel", "tijeras","papa","heavy", "lagarto", "spock"]
    opponent_choice = random.choice(choices)
    print(f'\nTu oponente jugó: {opponent_choice.capitalize()}.\n\n')
    return opponent_choice

def resolution(player_choice, opponent_choice):

    

    time.sleep(1)
    winmoves = {
        "roca": ["tijeras", "heavy"],
        "papel": ["roca", "spock"],
        "tijeras": ["papel", "lagarto","heavy"],
        "lagarto": ["papel", "spock"],
        "spock": ["tijeras", "roca"],
        "papa" : ["papel","roca"],
        "heavy" : ["lagarto","papa"],
        }

    if player_choice == opponent_choice:
        print("empatamos!\n")
        return

 
    elif opponent_choice in winmoves[player_choice]:
        print("ganaste!!\n")
    else:
        print("perdiste!\n")



print('''
Las tijeras mata el papel.
El papel mata la roca.
La roca mata el lagarto.
El lagarto mata a Spock.
Spock mata las tijeras.
Las tijeras mata el lagarto.
El lagarto mata el papel.
El papel mata a Spock.
Spock mata la roca.
La roca mata a las tijeras.
heavy mata papa.
heavy mata lagarto.
papa mata a papel.
papa mata a roca
roca mata heavy .
tijera mata heavy.
spock mata a papa.
\ 4_4  /
 \____/
 ''')
if __name__ == "__main__":
    print('bienvenido a roca papel o tijera ademas lagarto,y spock papa y heavy')

    # para seguir jugando
    retry_choice = None
    while retry_choice != "q":
        player_move = player()
        opponent_move = opponent()
        resolution(player_move, opponent_move)
        retry_choice = input('\nescribe "q" para detener "qq para seguir" ! ').lower()
```
